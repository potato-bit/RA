```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel('Live_Table_122.ods',header=3)
df1.head()
```

```python
years  = list(range(2002,2022))
df1_years = df1.iloc[:,7:].copy()
```

```python
df1_years.columns = years
```

```python
df = df1[['CurrentONS code','Lower and Single Tier Authority Data']].copy()
for x in range(len(years)):
    df[years[x]] = df1_years[years[x]]
```

```python
df.rename(columns={'CurrentONS code':'LA_code','Lower and Single Tier Authority Data':'LA'},inplace=True)
```

```python
df = df[~df.loc[:,'LA'].isnull()]
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df = pd.melt(df,id_vars=['LA_code','LA'],var_name='year',value_name='net dwellings')
```

```python
df['date'] = pd.to_datetime(df['year'].astype(str),format='%Y')
```

```python
df.to_excel('cleaned datasets/Net Dwellings by LA, 2001-2021.xlsx')
```
