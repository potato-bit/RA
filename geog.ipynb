```python
import pandas as pd
import numpy as np
import json
```

```python
with open('msoa.geojson') as json_data:
    msoa = json.load(json_data)
```

```python
df1 = pd.read_csv('geographies.csv')
```

```python
df1.dropna(axis=0,subset=['PCD','MSOA'],inplace=True)
```

```python
for x in df1['PCD']:
    print(x.strip(' '))
```

```python
msoa_cov = list(df1['MSOA'].unique())
len(msoa_cov)
```

```python
msoa_t = {'type': 'FeatureCollection', 'name': 'Middle_Layer_Super_Output_Areas_(December_2011)_Boundaries_Generalised_Clipped_(BGC)_EW_V3', 'crs': {'type': 'name', 'properties': {'name': 'urn:ogc:def:crs:OGC:1.3:CRS84'}}, 'features': []}
```

```python
index = []
for x in range(len(msoa['features'])):
    query = msoa['features'][x]['properties']['MSOA11CD']
    if query in msoa_cov:
        index.append(x)
        
```

```python
print(len(index),len(msoa_cov))
```

```python
for x in index:
    msoa_t['features'].append(msoa['features'][x])
```
