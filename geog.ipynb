```python
import pandas as pd
import numpy as np
import json
import geopandas as gpd
%matplotlib inline
```

```python
shapefile = gpd.read_file('PostalSector.shp')
print(shapefile)
```

```python
shapefile.head()
```

```python
postcodes = ['CV78','CV62','CV64','CV66','CV21','CV22','CV59','CV51','CV63','CV65','CV67','CV23','CV57','CV58','CV14','CV15','CV24','CV25',
'CV49','CV56','CV13','CV11','CV12','CV31','CV32','CV48','CV47','CV36','CV35','CV34','CV33']
```

```python
len(postcodes)
```

```python
shape_cov = shapefile[shapefile.loc[:,'StrSect'].isin(postcodes)].copy()
shape_cov = shape_cov[['StrSect','Locale','geometry']]
shape_cov
```

```python
df1 = pd.read_csv('geographies.csv')
```

```python
df1.dropna(axis=0,subset=['PCD','MSOA'],inplace=True)
```

```python
for x in df1.index:
    df1.loc[x,'StrSect'] = str(df1.loc[x,'PCD'])[0:5].replace(' ','')
```

```python
for x in df1.index:
    if df1.loc[x,'StrSect'] not in postcodes:
        df1.drop(x,axis=0,inplace=True)
```

```python
len(df1['StrSect'].unique())
```

```python
df2 = pd.read_excel('DC 28.4.22 Homefinder redacted.xlsx',sheet_name='Advertised Properties')
df2.head()
```

```python
df1.rename(columns={'PCD':'Postcode'},inplace=True)
```

```python
df2 = pd.merge(df2,df1[['Postcode','StrSect','LAT','LONG']],how='left',on='Postcode')
```

```python
df2.head()
```

```python
df2_agg1 = df2[['StrSect','Advert']].groupby('StrSect').count().rename(columns={'Advert':'Count'}).copy()
```

```python
df2_agg2 = df2[['StrSect','Number of Bedspaces','Bids','Offers','Refusals','Rejections']].groupby('StrSect').sum().copy()
```

```python
df2_agg = pd.merge(df2_agg1,df2_agg2,how='left',on='StrSect')
df2_agg.reset_index(inplace=True)
```

```python
df2_agg
```

```python
df = pd.merge(shape_cov,df2_agg,how='left',on='StrSect')
```

```python
gdf = gpd.GeoDataFrame(df)
```

```python
gdf.head()
```

```python
for i,x in enumerate(gdf.columns[3:]):
    gdf[f'{x}_p'] = gdf[x]/gdf[gdf.columns[3:]].sum()[i]*100
```

```python
gdf
```

```python
gdf.describe()
```

```python
merged_json = json.loads(gdf.to_json())
```

```python
json_data = json.dumps(merged_json)
```

```python
from bokeh.io import output_notebook, show, output_file
from bokeh.plotting import figure
from bokeh.models import GeoJSONDataSource, LinearColorMapper, ColorBar
from bokeh.palettes import brewer

geosource = GeoJSONDataSource(geojson=json_data)
palette = brewer['YlGnBu'][8]
palette = palette[::-1]
color_mapper = LinearColorMapper(palette = palette, low = 0, high = 15)
tick_labels = {'0': '0%', '2':'2%', '4':'4%', '6':'6%', '8':'8%', '10':'10%','12': '12%','14':'>14%'}
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=8,width = 500, height = 20,border_line_color=None,location = (0,0), 
                    orientation = 'horizontal',major_label_overrides = tick_labels)

p = figure(title='Share of Properties Advertised in Homefinder, 2021/22',plot_height = 600 , plot_width = 600, toolbar_location = None)
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.axis.visible = False


p.patches('xs','ys', source = geosource,fill_color = {'field' :'Bids_p', 'transform' : color_mapper},
          line_color = 'black', line_width = 0.25, fill_alpha = 1)
p.add_layout(color_bar,'below')

output_notebook()
show(p)



```
