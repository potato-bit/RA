```python
import pandas as pd
import numpy as np
```

```python
import os

directory = 'housing tax bands'
locations = []

for filename in os.listdir(directory):
    f = os.path.join(directory,filename)
    if os.path.isfile(f):
        locations.append(f)
```

```python
df = pd.read_csv(locations[-1])
df = df[df.loc[:,'geography']=='LAUA']
df.rename(columns={'ecode':'LA_code','area_name':'LA'},inplace=True)
df.drop(columns=['geography','ba_code'],inplace=True)
df['year'] = 2021
columns = df.columns
```

```python
data = []
for x,y in enumerate(locations):
    df = pd.read_csv(y,na_values=['..','-'])
    df = df[df.loc[:,'geography']=='LAUA']
    df.rename(columns={'ecode':'LA_code','area_name':'LA'},inplace=True)
    df.drop(columns=['geography','ba_code'],inplace=True)
    df.reset_index(drop=True,inplace=True)
    df['year'] = 1993 + x
    for k in df.values.tolist():
        data.append(k)
```

```python
df1 = pd.DataFrame(data)
df1.columns = columns
df1
```

```python
df1.to_excel('cleaned datasets/Number of Properties by Tax Band by LA, 1993-2021.xlsx')
```
