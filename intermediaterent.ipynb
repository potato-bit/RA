```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel('Live_Tables_1006-1009.ods',sheet_name='Live_Table_1007aC',header=3,na_values=['..'])
df1.head()
```

```python
df = df1[['CurrentONS code','Lower and Single Tier Authority Data']].copy()
df.columns = ['LA_code','LA']
```

```python
years = list(range(2004,2022))
df1_years = df1.iloc[:,3:]
```

```python
for x in range(len(years)):
    df[years[x]] = df1_years.iloc[:,x]
```

```python
df.dropna(axis=0,subset=['LA','LA_code'],inplace=True)
```

```python
df = pd.melt(df,id_vars=['LA_code','LA'],var_name='year',value_name='DC')
```

```python
df.to_excel('cleaned datasets/Intermediate Rent Dwellings by LA, 2003-2021.xlsx')
```
