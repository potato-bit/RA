```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows',200)
```

```python
df1 = pd.read_excel('Median House Prices by LA 1995-2021, Quarterly, Long.xlsx')
```

```python
df1.drop(index=34754,inplace=True)
#df1.dropna(0,subset=['median'],inplace=True)
print(df1.shape)
df1.head()
```

```python
def quarterer(x):
    quarter = ''
    month = x.split(' ')[0]
    year = x.split(' ')[-1]
    if month == 'Mar':
        quarter = ' Q1'
    elif month == 'Jun':
        quarter = ' Q2'
    elif month == 'Sep':
        quarter = ' Q3'
    elif month == 'Dec':
        quarter = ' Q4'
    result = year + quarter
    return result
```

```python
df1['quarter'] = df1['quarter'].map(quarterer)   
```

```python
df1.shape
```

```python
df1 = df1[df1.loc[:,'quarter']!='...109']
df1.shape
```

```python
df2 = pd.read_csv('CPIH.csv')
df2 = df2[72:176]
df2.rename(columns={'Title':'quarter','CPIH INDEX 00: ALL ITEMS 2015=100':'cpih'},inplace=True)
df2['cpih'] = df2['cpih'].apply(pd.to_numeric)
df2.reset_index(inplace=True,drop=True)
df2.head()
```

```python
base = float(df2['cpih'][df2.loc[:,'quarter']=='2015 Q2'])
```

```python
df1.head()
```

```python
df1 = df1.merge(df2,on='quarter',how='inner')
```

```python
df1.sort_values(['LA_code','quarter'],ignore_index=True,inplace=True)
```

```python
df1['real_median'] = (df1['median']*base)/df1['cpih']
df1.head(200)
```

```python
df1.to_excel('2015 Real Median House Price by LA 1995-2021, Quarterly.xlsx')
```
