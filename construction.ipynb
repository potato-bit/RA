```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows',100)
```

```python
df1 = pd.read_excel('construction index, 2015=100.xlsx')
df1.head()
```

```python
df1 = df1[df1.index!=12]
```

```python
df1 = pd.melt(df1,id_vars=['month'],var_name='year',value_name='indice')
```

```python
df1['date'] = pd.to_datetime(df1['year'].astype(str) + df1['month'].astype(str), format='%Y%m')
```

```python
df1.to_excel('cleaned datasets/construction_long.xlsx')
```
